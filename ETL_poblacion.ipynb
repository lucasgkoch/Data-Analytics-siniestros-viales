```python
import pandas as pd
```

```python
#Se carga el archivo csv
df_poblacion=pd.read_csv('raw_data\informacion-censal-por-radio-2010.csv')
df_poblacion.head()
```

```python
#Se extraen las columnas necesarias
df_poblacion=df_poblacion[['COMUNA','TOTAL_POB']]
df_poblacion.columns=df_poblacion.columns.str.lower()
```

```python
#Se agrupa por comunas
df_poblacion=df_poblacion.groupby('comuna')['total_pob'].sum().reset_index()
df_poblacion
```

```python
#Se exporta el archivo csv
df_poblacion.to_csv('processed_data/poblacion.csv',index=False)
```
