```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Se carga el dataset de homicidios_hechos
df_homicidios_hechos=pd.read_excel('raw_data\homicidios.xlsx', sheet_name='HECHOS', engine='openpyxl')
df_homicidios_hechos.head()
```

Normalizacion en nombres de columnas

```python
#Se pasan a minusculas los nombres de las columnas
df_homicidios_hechos.columns = df_homicidios_hechos.columns.str.lower()
#Se eliminan espacios y tildes
df_homicidios_hechos.rename(columns={'dirección normalizada': 'direccion_normalizada','xy (caba)': 'xy(caba)','pos x': 'longitud','pos y': 'latitud'}, inplace=True)

df_homicidios_hechos.head()
```

```python
df_homicidios_hechos['fecha']=pd.to_datetime(df_homicidios_hechos['fecha'])
```

```python
df_homicidios_hechos.dtypes
```

```python
df_homicidios_hechos.drop(columns=['lugar_del_hecho','calle','altura','cruce','direccion_normalizada','xy(caba)'],inplace=True)
```

Tratamiento de valores faltantes y duplicados

```python
#Se identifiacn valores faltantes por columna
df_homicidios_hechos.isnull().sum()
```

```python
#Se visualizan los valores unicos de la columna comunas
df_homicidios_hechos['comuna'].unique()
```

```python
#Se considera los registros con comuna=0 como valor faltante (la comuna 0 no existe)
df_homicidios_hechos[df_homicidios_hechos['comuna'] == 0]
```

```python
#Se elimina el registro ya que no contiene informacion que es necesria para nuestro analisis
df_homicidios_hechos = df_homicidios_hechos[df_homicidios_hechos['comuna'] != 0]
```

```python
#Se verifica la existencia de registros duplicados
df_homicidios_hechos[df_homicidios_hechos.duplicated(subset=df_homicidios_hechos.columns.difference(['id']))]
```

```python
df_homicidios_hechos['victima'].unique()
```

```python
df_homicidios_hechos[df_homicidios_hechos['victima']=='OBJETO FIJO']
```

```python
#Se detectan outliers y se eliminan
df_homicidios_hechos=df_homicidios_hechos[df_homicidios_hechos['victima']!='OBJETO FIJO']
```

```python
#Se realiza un histograma de la distribucion de las victimas de los homicidios
plt.hist(df_homicidios_hechos['victima'], bins=24, edgecolor='black')
plt.title('Distribución de victimas de Homicidios')
plt.xlabel('Hora del día')
plt.ylabel('Frecuencia')
plt.xticks(rotation='vertical')
plt.show()
```

```python
#Con fines de poder lograr graficos mas explicativos para nuestro analisis, dejaremos solo las categorias mas relevantes y al resto las agruparemos dentro de 'OTROS'

def asignar_categoria(victima):
    categorias_validas = ['MOTO', 'PEATON', 'AUTO', 'BICICLETA']
    return victima if victima in categorias_validas else 'OTROS'

df_homicidios_hechos['victima'] = df_homicidios_hechos['victima'].apply(asignar_categoria)
```

Se guarda el dataset limpio en formato csv

```python
df_homicidios_hechos.to_csv('processed_data/homicidios_hechos.csv',index=False)
```
